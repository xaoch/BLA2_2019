## Introduction to Predictive Models

During this tutorial we will visit the simplier of predictive models, Linear Regression.

Now we will start to analyze how the different variables determine the probability of admision.  First we will try to visualize the multidimensional space of the variables in 2 dimensions using <a href="https://en.wikipedia.org/wiki/Principal_component_analysis">Principal Component Analysis </a>.

For this, first we need to load the library sklearn that provide several statistic and machine learning functions.

```python
from sklearn.decomposition import PCA
```

<font color="darkblue">**Explanation:**
<ul>
<li>The sklearn have several modules, the Principal Component Analysis function is inside the decomposition module.
<li>To load a single function from a module, we use the "from" ... "import" ... command.
</ul>
</font>

We now select only the columns from the dataset that have the independant variables that we want to use.  All but the Serial No. and the Chance of Admission.

```python
X = df[df.columns.difference(['Chance of Admit ', 'Serial No.'])]
```

<font color="darkblue">**Explanation:**
<ul>
<li>To select all but certain columns, we use the difference function form the list of columns of the dataframe.  We send as parameter the columns that we want to remove.
<li>We use the result of this call as the selection parameter of the datafram.
</ul>
</font>

Now we reduce the dimensionality of the dataframe to 2 dimensions using PCA

```python
pca = PCA(n_components=2)
pca_fit = pca.fit(X)
print(pca.explained_variance_ratio_)
```

<font color="darkblue">**Explanation:**
<ul>
<li>First, we configure the PCA process, we create the "pca" object as a PCA calulation with 2 components ("n_components" parameter).
<li>In the second line, we fit the data to calculate this two components.
<li>In the third line, we call to the function "explained_variance_ration" to know what percentage of the variance of the original data is explained by each component.  In this case, the first component explain almost 93% of the original variance, while the second explains a 6%.  This two components are the combinations of the different original variables.
</ul>
</font>

Now, we convert the original data (with its 7 independent variables) into a dataset with just two variables

```python
principalComponents = pca.transform(X)
principalDf = pd.DataFrame(data=principalComponents, columns = ['principal component 1', 'principal component 2'])
```

<font color="darkblue">**Explanation:**
<ul>
<li>In the first line, we project the multidimensional space into the two principal components (pca.transform) and we store the result in "principalComponents"
<li>In the second line we create a new dataframe using the "principalComponents" data.  We can set the name of the colums of this new dataframe.
</ul>
</font>

Now we create a scatterplot where each axis is one of the components.

```python
Y=df['Chance of Admit ']

plt.scatter(x=principalDf['principal component 1'],y=principalDf['principal component 2'], c=Y, cmap="BuPu_r" )
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2 Dimensional representation of the data')
plt.rcParams['figure.figsize'] = [12, 10]
plt.show()
```

<font color="darkblue">**Explanation:**
<ul>
<li>In the first line, we define Y as the column Chance to be Admitted.
<li>In the second line, we create a new scatterplot where the x is the principal compoenent 1, y is the principal component 2, the color ("c") will be determined by the change to be admitted, and the pallette of colors will be from blue to purple.
<li>The third to fifht lines add the labels
<li>The sixth line specify the size of the image.
<li>The seventh line present the plot.
<li>Now is easy to see why the first component explain 93% of the variation.  The color of the points change consistently from left to right.
</ul>
</font>

### Linear Regression

We are not interesting in understanding or exploring the dataset, we are interested in creating a model that we can use to predict the chance of student based on the independend variables.  

The easiest model that we can create is a linear regression.  This model consist in estimating the parameters (m's and b) of the linear function that best fit the data (Y=m1X1+m2X2+m3X3+...+b).  Once we have the m's and the b, we can plug the values for the independen variables (Xs) and obtain a value for the dependandt variable (Y), in this case, the probability of being accepted.

Let's start.

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

<font color="darkblue">**Explanation:**
<ul>
<li>We import several functions:
<li>The train_test_split to divide the data into a training and a test set.
<li>The LinearRegression to compute the linear regression
<li>The metrics to calculate how good was is our model
</ul>
</font>

When we train a model, we do not want to use all our data to train it.  If we do it, we can overfit our model, that means that it is able to almost perfectly predict the train data, but we do not have an idea how it will behave with new data.  To avoid that, a estrategy is to separate the data into a train and a test set.  We train the model with the train dataset, and we establish the accuracy of our model with the test set.  

Let's proceed to divide our data.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

<font color="darkblue">**Explanation:**
<ul>
<li>We use the train_test_split function to divide the data into a 80% train and 20% test (test_size parameter).  Two variables (X - independant variables and Y - dependant variable) also enter the function.
<li>The result of the function are 4 dataframes (X_train, X_test, Y_train and Y_test).
</ul>
</font>

After splitting the data into training and testing sets, finally, the time is to train our model. For that, we need to instanciate our LinearRegression class, and call the fit() method along with our training data.

```python
regressor = LinearRegression()  
regressor.fit(X_train, Y_train) 
```

<font color="darkblue">**Explanation:**
<ul>
<li>In the first line we instanciate the LinearRegression object.
<li>In the second line, we fit the linear fucntion to the train data.
</ul>
</font>

As we have discussed that the linear regression model basically finds the best value for the intercept and slope, which results in a line that best fits the data. To see the value of the intercept and slope calculated by the linear regression algorithm for our dataset, execute the following code.

```python
print("Intercept:",regressor.intercept_)

coeff_df = pd.DataFrame(regressor.coef_, X.columns, columns=['Coefficient'])  
coeff_df
```

Now that we have trained our algorithm, it’s time to make some predictions. To do so, we will use our test data and see how accurately our algorithm predicts the percentage score. To make predictions on the test data, execute the following script:



```python
Y_pred = regressor.predict(X_test)
```

```python
X_case =pd.DataFrame({'CGPA':[1],'GRE Score':[5],'LOR':[1],'Research':[1],'SOP':[1],'TOEFL Score':[1],'University Rating':[1]})
Y_case=regressor.predict(X_case)
Y_case[0]
```

Now compare the actual output values for X_test with the predicted values, execute the following script:

```python
prediction = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})
prediction.head()
```

We can also visualize comparison result as a bar graph using the below script:
    
Note: As the number of records is large, for representation we are using only 25 students.

```python
prediction25 = prediction.head(25)
prediction25.plot(kind='bar',figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

<font color="darkblue">**Explanation:**
<ul>
<li>In the first line, we select only the first 25 students.
<li>In the second line, we use the property of the dataframes in Python that allow to send them to be ploted.  We specify that we want a bar plot. 
<li>The following lines specify the grid division for the graph. 
</ul>
</font>

Though our model is not very precise, the predicted percentages are close to the actual ones.

Let's plot our straight line with the test data.

```python
plt.scatter(Y_test, Y_pred,  color='gray')
ident = [min(Y_test), max(Y_test)]
plt.plot(ident,ident, color="red")
plt.show()
```

<font color="darkblue">**Explanation:**
<ul>
<li>First we plot the predicted vs. real value 
<li>In the second line, we use the property of the dataframes in Python that allow to send them to be ploted.  We specify that we want a bar plot. 
<li>The next line create the values of an identity line from the minimum observation to the maximum observation.
<li>The fourth line, plots that line into the graph.
</ul>
</font>

Finally, we will measure the quality of our model.  In this case, Linear Regression, the quality is measured by the distance between the predicted and real values in the Test set.

We make the following calculations using the metrics package.

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, Y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, Y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
```

This result tell us that the mean squared error (the average difference between the prediction and the real value) is 0.055.  This around a 10% of the average value (0.5).

Another common way to measurethe quality of a Linear Regression model is R2.  We use the score property of the model

```python
regressor.score(X_test,Y_test)
```

That means that our model explain 84% of the variance in the test data.  It is actually a good model.

## Creating a Python Dashboard

Now we will create a dashboard that could generate a prediction for any given combination of values.

We will use dash, a Python library to create dashboards.  This library does not come with Anaconda. We need to install it in our computers.

Dash apps are composed of two parts. The first part is the "layout" of the app and it describes what the application looks like. 

The following is a test code that you can run in your machine.  Open Spyder, and paste this code in the editor.  Then run button.

<pre>
# -*- coding: utf-8 -*-
import dash
import dash_core_components as dcc
import dash_html_components as html

external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

app = dash.Dash(__name__, external_stylesheets=external_stylesheets)
server = app.server
app.title="Test Dashboard"

app.layout = html.Div(children=[
    html.H1(children='Hello Dash'),

    html.Div(children='''
        Dash: A web application framework for Python.
    '''),

    dcc.Graph(
        id='example-graph',
        figure={
            'data': [
                {'x': [1, 2, 3], 'y': [4, 1, 2], 'type': 'bar', 'name': 'SF'},
                {'x': [1, 2, 3], 'y': [2, 4, 5], 'type': 'bar', 'name': u'Montréal'},
            ],
            'layout': {
                'title': 'Dash Data Visualization'
            }
        }
    )
])

if __name__ == '__main__':
    app.run_server()
</pre>

<font color="darkblue">**Explanation:**
<ul>
<li>The layout is composed of a tree of "components" like html.Div and dcc.Graph.
<li>The dash_html_components library has a component for every HTML tag. The html.H1(children='Hello Dash') component generates a \Hello Dash HTML element in your application.
<li>Not all components are pure HTML. The dash_core_components describe higher-level components that are interactive and are generated with JavaScript, HTML, and CSS through the React.js library.
<li>The children property is special. By convention, it's always the first attribute which means that you can omit it: html.H1(children='Hello Dash') is the same as html.H1('Hello Dash'). Also, it can contain a string, a number, a single component, or a list of components.
</ul>
</font>
